```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
%matplotlib inline


x1 = np.linspace(0, 10, 100)


# create a plot figure
fig = plt.figure()

plt.plot(x1, np.sin(x1), '-')
plt.plot(x1, np.cos(x1), '--');
```

```python
# create a plot figure
plt.figure(figsize=(10, 10))


# create the first of two panels and set current axis
plt.subplot(2, 1, 1)   # (rows, columns, panel number)
plt.plot(x1, np.sin(x1))


# create the second of two panels and set current axis
plt.subplot(2, 1, 2)   # (rows, columns, panel number)
plt.plot(x1, np.cos(x1));
```

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()

plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t**1.5, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
x = np.linspace(1, 100, 2)
plt.plot(x, x**2, 'r--', label='Test')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

Object oriented APIs


```python
figure = plt.figure()

x1 = np.linspace(1, 5)
y1 = np.sin(x1**2)

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(x1, y1, 'r')
```

```python
figure = plt.figure()
x1 = np.linspace(0.1, 2, num=25)

axes1 = figure.add_subplot(2, 2, 1)
axes1.plot(x1, np.log10(x1))

axes2 = figure.add_subplot(2, 2, 2)
axes2.plot(x1, np.log(x1), 'r')

axes3 = figure.add_subplot(2, 2, 3)
axes3.plot(x1, np.sin(x1), 'g')

axes4 = figure.add_subplot(2, 2, 4)
axes4.plot(x1, np.cos(x1), 'b')
```
